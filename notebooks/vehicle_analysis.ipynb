```python
# Import pandas and statsmodels
import pandas as pd
import statsmodels.api as sm

df_auto_clean = df_auto[["brand_model","km","age","price"]].dropna()


# Group the dataframe by model
grouped = df_auto_clean.groupby('brand_model')

# For each model, perform linear regression and print the equation
for name, group in grouped:
    # Define the dependent and independent variables
    y = group['price']
    X = group[['km', 'age']]
    # Add a constant term to the independent variables
    X = sm.add_constant(X)
    # Fit the linear regression model
    model = sm.OLS(y, X).fit()
    # Get the coefficients
    coef = model.params
    # Print the equation
    print(f"Model {name}: price = {coef[0]:.2f} + {coef[1]:.2f} * km + {coef[2]:.2f} * age")
```
